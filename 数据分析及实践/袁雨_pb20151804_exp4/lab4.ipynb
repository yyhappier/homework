```python
# Importing the libraries
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
from sklearn import datasets, linear_model
from sklearn.model_selection import cross_validate
from sklearn.metrics import mean_squared_error, r2_score
from sklearn import preprocessing
from sklearn.model_selection import cross_val_score


import xgboost as xgb
```

```python
# 数据导入
initialdata = pd.read_csv("D:\\jupyterlab\\lab4\\pica2015_new.csv",na_values=' ')
df = pd.DataFrame(initialdata)
# 均值填充缺失值
pd.set_option('display.max_rows',None)
pd.set_option('display.max_columns',None)
for column in list(df.columns[df.isnull().sum() > 0]):
    mean_val = df[column].mean()
    df[column].fillna(mean_val, inplace=True)
print(df.isna().sum())
```

```python
# 计算pearsonr系数
from scipy.stats.stats import pearsonr
pea=[[] for i in range(430)]
for i in range(19,430):
    temp=df.iloc[:, i]
    if temp.dtypes!='object':
        pea[i]=[df.iloc[0:0,i].name,pearsonr(temp,df['MATH'])[0]]
        print(pea[i])
```

```python
for i in range(19,430):
    if pea[i][1]<0:
        pea[i][1]=-1*pea[i][1]
peadf=pd.DataFrame(pea,columns=['NAME','PEARSONR'])
# peadf.dropna(inplace=True)
# peadfabs=abs(peadf)
peadf.sort_values(by='PEARSONR', inplace=True,ascending = False)  #  inplace=True，表示直接替换原有df对象
print(peadf)
```

```python
# Use feature PV
X_PV=df.iloc[:,328:428]
# Use feature ST
X_ST=df.iloc[:,21:245]
# Use feature ST+PV+REPEAT
X_STPVRE=pd.concat([X_ST,X_PV,df['REPEAT']], axis=1)
print(X_STPVRE.head())
# Use all features
X_all3=df.iloc[:,19:429]
# LOG
LOGX_PV=np.log(X_PV)
LOGX=pd.concat([LOGX_PV, X_ST,df['REPEAT']], axis=1)
print(LOGX.head())
```

```python
# Y
Y=df.iloc[:,-1]
```

```python
# 预处理
Normalizer = preprocessing.Normalizer().fit_transform(X_STPVRE)
MinMaxScaler=preprocessing.MinMaxScaler().fit_transform(X_STPVRE)
StandardScaler=preprocessing.StandardScaler().fit_transform(X_STPVRE)
```

```python
# LinearRegression
regr = linear_model.LinearRegression()
score = cross_val_score(regr,Normalizer, Y, cv=5,scoring='neg_mean_squared_error')
print(score)
print(np.mean(score))
```

```python
# Ridge
from sklearn.linear_model import Ridge
ridge = Ridge(alpha=1.5)
scores = cross_val_score(ridge, MinMaxScaler, Y, cv=5, scoring='neg_mean_squared_error')
print(scores)
print(np.mean(scores))
```

```python
# BayesianRidge
regr = linear_model.BayesianRidge(tol=1e-4)
score=cross_val_score(regr,MinMaxScaler, Y, cv=5,scoring='neg_mean_squared_error')
print(score)
print(np.mean(score))
```

```python
# TweedieRegressor
regr = linear_model.TweedieRegressor(power=0, alpha=1, link='log')
scores = cross_validate(regr,StandardScaler, Y, cv=5,
                         scoring=('r2','neg_mean_squared_error'),
                         return_train_score=True)
score=scores['test_neg_mean_squared_error']
print(score)
print(np.mean(score))
```

```python
# LinearSVR
from sklearn import svm
regr = svm.LinearSVR(dual=False,tol=1e-5,C=10,loss='squared_epsilon_insensitive')
score=cross_val_score(regr,StandardScaler, Y, cv=5,scoring='neg_mean_squared_error')
print(score)
print(np.mean(score))
```

```python
# KNeighborsRegressor
from sklearn import neighbors
regr = neighbors.KNeighborsRegressor(n_neighbors=50)
score=cross_val_score(regr,Normalizer, Y, cv=5,scoring='neg_mean_squared_error')
print(score)
print(np.mean(score))
```

```python
# DecisionTreeRegressor
from sklearn.tree import DecisionTreeRegressor
regr = DecisionTreeRegressor(max_depth=5)
score=cross_val_score(regr,StandardScaler, Y, cv=5,scoring='neg_mean_squared_error')
print(score)
print(np.mean(score))
```
